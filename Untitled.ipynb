```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your dataset is stored in a CSV file named "population_data.csv"

df = pd.read_csv("C:\\Users\\pasup\\Untitled Folder 1\\API_SP.POP.TOTL_DS2_en_csv_v2_6298256.csv", skiprows=4, sep=',')

# Choose the year you want to visualize (e.g., 2019)
year_to_visualize = '2019'

# Filter out unnecessary columns
df_filtered = df[['Country Name', year_to_visualize]]

# Drop rows with missing values
df_filtered = df_filtered.dropna()

# Sort the data by population in descending order
df_filtered = df_filtered.sort_values(by=year_to_visualize, ascending=False)

# Choose the top N countries for better visualization
top_n = 10
df_top = df_filtered.head(top_n)

# Create a bar chart
plt.figure(figsize=(12, 8))
plt.bar(df_top['Country Name'], df_top[year_to_visualize], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Population in {}'.format(year_to_visualize))
plt.title('Top {} Countries by Population in {}'.format(top_n, year_to_visualize))
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Display the chart
plt.show()
```
